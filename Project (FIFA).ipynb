# Feature Engineering

```python
project.drop(["Unnamed: 0","Photo","Flag","Club Logo","Real Face"], inplace=True, axis=1)
```

```python
project['Body Type'].unique()
```

```python
project['Body Type'].replace({'Messi':'Lean','C. Ronaldo':'Stocky','Neymar':'Lean','Courtois':'Normal','PLAYER_BODY_TYPE_25':'Lean','Shaqiri':'Lean','Akinfenwa':'Lean'}, inplace=True)
```

```python
project.dropna(thresh=16000, inplace=True, axis=1)
```

```python
for i in project.select_dtypes(exclude = ['object']):
    project[i].fillna(project[i].median(), inplace=True)
```

```python
project['Contract Valid Until']=pd.DatetimeIndex(project['Contract Valid Until']).year
```

```python
columns_01 = ['Club','Preferred Foot','Work Rate','Body Type','Position','Joined','Contract Valid Until']
for i in columns_01:
    project[i].fillna('Unknown', inplace=True)
```

```python
import re
r = re.compile(r"([0-9]+)'([0-9]{1,2})")
def get_inches(el):
    m = r.match(el)
    if m == None:
        return float('NaN')
    else:
        return (int(m.group(1))*12 + float(m.group(2)))*2.54
```

```python
project['Height'] = project['Height'].apply(str)
```

```python
project['Height'] = project['Height'].apply(get_inches)
```

```python
project['Height'].fillna(project['Height'].median(), inplace=True)
```

```python
r = re.compile(r"([0-9]+)lbs")
def get_kg(el):
    m = r.match(el)
    if m == None:
        return float('NaN')
    else:
        return (int(m.group(1))*0.45359)
```

```python
project['Weight'] = project['Weight'].apply(str)
```

```python
project['Weight'] = project['Weight'].apply(get_kg)
```

```python
project['Weight'].fillna(project['Weight'].median(), inplace=True)
```

```python
columns_02 = ['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM',
             'RDM','RWB','LB','LCB','CB','RCB','RB']

for i in columns_02:
    new_column = project[i].str.split("+", n=1, expand=True)
    project[i] = new_column[0]
    project[i+"(+)"] = new_column[1]
```

```python
columns_03 = ['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB',
          'LS(+)','ST(+)','RS(+)','LW(+)','LF(+)','CF(+)','RF(+)','RW(+)','LAM(+)','CAM(+)','RAM(+)','LM(+)','LCM(+)','CM(+)','RCM(+)','RM(+)','LWB(+)',
           'LDM(+)','CDM(+)','RDM(+)','RWB(+)','LB(+)','LCB(+)','CB(+)','RCB(+)','RB(+)']
for i in columns_03:
    project[i].fillna(project[i].median(), inplace=True)
```

```python
project['Release Clause_copy'] = project['Release Clause']
```

```python
import re
t = re.compile(r"€(([0-9]*[.])?[0-9]+)K")
r = re.compile(r"€(([0-9]*[.])?[0-9]+)M")
def get_money(el):
    n = t.match(el)
    m = r.match(el)
    if el[-1]=='K':
        return (float(n.group(1))*1000)
    elif el[-1]=='M':
        return (float(m.group(1))*1000000)
    else:
        return float('NaN')
```

```python
project['Release Clause_copy'] = project['Release Clause_copy'].apply(str)
```

```python
project['Release Clause_copy'] = project['Release Clause_copy'].apply(get_money)
```

```python
project['Release Clause_copy'].fillna(project['Release Clause_copy'].median(), inplace=True)
```

```python
project['Release Clause_copy'] = project['Release Clause_copy'].apply(int)
```

```python
project['Release Clause'].fillna('€1.1M', inplace=True)
```

```python
list_back = ['CB','LB','RB','LCB','RCB','RWB','LWB']
list_forward = ['ST','LW','RW','LS','RS','CF','RF','LF']
def new_position(value):
    if (value == 'GK'):
        return 'Goalkeeper'
    elif (value in list_back):
        return 'Defender'
    elif (value in list_forward):
        return 'Forward'
    else:
        return 'Midfielder'

project['new_position'] = project['Position'].apply(new_position)
```

# Modelling

```python
X = project.drop(['Potential','ID','Name','Nationality','Club','Value','Wage','new_position','Joined','Contract Valid Until','Release Clause'], axis=1)
y = project['Potential']
```

```python
from sklearn.model_selection import KFold

kf  = KFold(n_splits = 10)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler()

X_train_norm = norm.fit(X_train).transform(X_train)

X_test_norm = norm.transform(X_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

model_1 = LinearRegression()

model_1.fit(X_train_norm, y_train)

y_pred_1 = model_1.predict(X_test_norm)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print("MAE:", round(mean_absolute_error(y_test,y_pred_1),2))
print("MSE:", round(mean_squared_error(y_test,y_pred_1),2))
print("RMSE:", round(np.sqrt(mean_squared_error(y_test,y_pred_1)),2))
print("RMSLE:", round(np.log(np.sqrt(mean_squared_error(y_test,y_pred_1))),2))
print("R2:", round(r2_score(y_test,y_pred_1),2))
```

```python
print("Intercept:",model_1.intercept_)
print("Slope:",model_1.coef_)
```

```python
plt.figure(figsize=(18,10))
sns.regplot(y_pred_1, y_test, scatter_kws={'color':'red','edgecolor':'blue'}, line_kws={'color':'white', 'alpha':0.5})
plt.xlabel('Predictions', fontsize=20)
plt.ylabel('Overall', fontsize=18)
plt.title("Linear Prediction of Players' Potential", fontsize=20)
plt.show()
```

# SVR

```python
from sklearn.svm import SVR

model_2 = SVR(kernel='linear', C=9.0)

model_2.fit(X_train_norm, y_train)

y_pred_2 = model_2.predict(X_test_norm)
```

```python
print("MAE:", round(mean_absolute_error(y_test,y_pred_2),2))
print("MSE:", round(mean_squared_error(y_test,y_pred_2),2))
print("RMSE:", round(np.sqrt(mean_squared_error(y_test,y_pred_2)),2))
print("RMSLE:", round(np.log(np.sqrt(mean_squared_error(y_test,y_pred_2))),2))
print("R2:", round(r2_score(y_test,y_pred_2),2))
```

```python
print("Intercept:",model_2.intercept_)
print("Slope:",model_2.coef_)
```

```python
plt.figure(figsize=(18,10))
sns.regplot(y_pred_2, y_test, scatter_kws={'color':'red','edgecolor':'blue'}, line_kws={'color':'white', 'alpha':0.5})
plt.xlabel('Predictions', fontsize=20)
plt.ylabel('Overall', fontsize=18)
plt.title("SVR Prediction of Players' Potential", fontsize=20)
plt.show()
```
